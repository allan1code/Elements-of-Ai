```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming num_customers is defined 
num_customers = 100

# Generate sample data
customer_ids = range(1, num_customers + 1)
ages = np.random.randint(18, 70, num_customers)
genders = np.random.choice(['Male', 'Female'], num_customers)
customer_locations = np.random.choice(['City', 'Suburb', 'Rural'], num_customers)
registration_dates = pd.date_range('20220101', periods=num_customers)
purchase_amounts = np.random.normal(100, 20, num_customers)
num_orders = np.random.randint(1, 10, num_customers)
avg_order_values = np.random.normal(80, 10, num_customers)
last_purchase_dates = pd.date_range('20220101', periods=num_customers)
satisfaction_scores = np.random.randint(1, 11, num_customers)
customer_engagement = np.random.choice(['High', 'Medium', 'Low'], num_customers)
product_preferences = np.random.choice(['A', 'B', 'C'], num_customers)
subscription_lengths = np.random.randint(1, 13, num_customers)
subscription_types = np.random.choice(['Monthly', 'Yearly'], num_customers)
subscription_start_dates = pd.date_range('20220101', periods=num_customers)
churned = np.random.choice(['Yes', 'No'], num_customers)
churn_dates = pd.date_range('20220101', periods=num_customers)
churn_reasons = np.random.choice(['Price', 'Service', 'Product Quality', 'Others'], num_customers)

# Marketing Interactions
email_responses = np.random.randint(0, 2, num_customers)
ad_click_rates = np.random.uniform(0, 1, num_customers)
coupon_usage = np.random.randint(0, 5, num_customers)

# Feedback and Reviews
product_ratings = np.random.randint(1, 6, num_customers)
customer_reviews = ['Good' if rating >= 4 else 'Average' if rating == 3 else 'Bad' for rating in product_ratings]
nps_scores = np.random.randint(0, 11, num_customers)

# Financial Data
total_revenue = np.random.normal(5000, 1000, num_customers)
profit_margin = np.random.uniform(0.1, 0.5, num_customers)

# Calculate Lifetime Customer Value
lifetime_customer_value = total_revenue * subscription_lengths * profit_margin

# Create DataFrame
data = {
    'Customer_ID': customer_ids,
    'Age': ages,
    'Gender': genders,
    'Location': customer_locations,
    'Registration_Date': registration_dates,
    'Purchase_Amount': purchase_amounts,
    'Number_of_Orders': num_orders,
    'Average_Order_Value': avg_order_values,
    'Last_Purchase_Date': last_purchase_dates,
    'Satisfaction_Score': satisfaction_scores,
    'Customer_Engagement': customer_engagement,
    'Product_Category_Preference': product_preferences,
    'Subscription_Length_Months': subscription_lengths,
    'Subscription_Type': subscription_types,
    'Subscription_Start_Date': subscription_start_dates,
    'Churned': churned,
    'Churn_Date': churn_dates,
    'Churn_Reason': churn_reasons,
    'Email_Responses': email_responses,
    'Ad_Click_Rate': ad_click_rates,
    'Coupon_Usage': coupon_usage,
    'Product_Rating': product_ratings,
    'Customer_Review': customer_reviews,
    'NPS_Score': nps_scores,
    'Total_Revenue': total_revenue,
    'Profit_Margin': profit_margin,
    'Lifetime_Customer_Value': lifetime_customer_value
}

df = pd.DataFrame(data)

# Save data to CSV
df.to_csv(Ecommerce_Dataset_With_LCV.csv', index=False)

# Visualization and Analysis
# Example: Pairplot using Seaborn
sns.pairplot(df[['Age', 'Purchase_Amount', 'Satisfaction_Score', 'Lifetime_Customer_Value']])
plt.show()
```
